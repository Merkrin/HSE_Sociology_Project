```python
import pandas as pd
import scipy.stats as stats
```

```python
time_column = 'Отметка времени'
sex_column = 'Ваш пол'
age_column = 'Сколько вам полных лет'
music_column = 'Как часто вы слушаете музыку по собственному желанию'
sweets_column = 'Как часто вы едите конфеты по собственному желанию'
mark_column = 'Какая ваша предпочитаемая марка конфет'
type_column = 'Какой тип конфет вы предпочитаете'
genre_column = 'Какой ваш предпочитаемый жанр музыки'
singer_column = 'Какой ваш любимый исполнитель'
```

# Связь любимых марок конфет с любимыми жанрами
# Разбиение строк по множественным ответам

```python
for column in [mark_column, genre_column]:
    dataframe = dataframe.explode(column)
    
dataframe
```

## Таблица сопряжённости

```python
alpha = 0.05

data_crosstab = pd.crosstab(dataframe[mark_column], dataframe[genre_column], margins=True, margins_name="Total")

data_crosstab
```

## Тест хи-квадрат
### P-value approach

```python
chi_square = 0

rows = dataframe[mark_column].unique()
columns = dataframe[genre_column].unique()

for i in columns:
    for j in rows:
        O = data_crosstab[i][j]
        E = data_crosstab[i]['Total'] * data_crosstab['Total'][j] / data_crosstab['Total']['Total']
        chi_square += (O-E)**2/E
```

```python
print("Approach 1: The p-value approach to hypothesis testing in the decision rule")
p_value = 1 - stats.norm.cdf(chi_square, (len(rows)-1)*(len(columns)-1))
conclusion = "Failed to reject the null hypothesis."
if p_value <= alpha:
    conclusion = "Null Hypothesis is rejected."
        
print("chisquare-score is:", chi_square, " and p value is:", p_value)
print(conclusion)
```

### Critical value approach

```python
print("Approach 2: The critical value approach to hypothesis testing in the decision rule")
critical_value = stats.chi2.ppf(1-alpha, (len(rows)-1)*(len(columns)-1))
conclusion = "Failed to reject the null hypothesis."
if chi_square > critical_value:
    conclusion = "Null Hypothesis is rejected."
        
print("chisquare-score is:", chi_square, " and critical value is:", critical_value)
print(conclusion)
```
